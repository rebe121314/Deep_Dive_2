```python
var('x,y,l')
F(x,y) = 3*x+4*y
G1(x,y,z) = sqrt(x)+ sqrt(y)
eqs = [F.gradient()[i] == l*G1.gradient()[i] for i in range(2)]
eqs.append(G1==7)
sol=solve(eqs,x,y,l)
print("Solution for the system",sol)

extrema=[F(sol[1].rhs(),sol[2].rhs()).n() for sol in sol]
print("Critical point",extrema)

#print(F(16,9)
```
